```python
import matplotlib.pyplot as plt

# 데이터 정의
samples = {
    'A': {'width': 5.5, 'lightness': 3.2, 'label': 'Salmon'},
    'B': {'width': 6.1, 'lightness': 2.8, 'label': 'Sea Bass'},
    'C': {'width': 5.9, 'lightness': 3.5, 'label': 'Salmon'},
    'D': {'width': 6.8, 'lightness': 2.5, 'label': 'Sea Bass'},
}

# 각 클래스별 좌표 나누기
salmon_x = []
salmon_y = []
seabass_x = []
seabass_y = []

for sample in samples.values():
    x = sample['lightness']
    y = sample['width']
    if sample['label'] == 'Salmon':
        salmon_x.append(x)
        salmon_y.append(y)
    else:
        seabass_x.append(x)
        seabass_y.append(y)

# 시각화
plt.figure(figsize=(6, 5))
plt.scatter(salmon_x, salmon_y, color='orange', label='Salmon (●)', marker='o', s=100)
plt.scatter(seabass_x, seabass_y, color='blue', label='Sea Bass (×)', marker='x', s=100)

plt.xlabel("Lightness")
plt.ylabel("Width (cm)")
plt.title("Fish Classification: Lightness vs Width")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
